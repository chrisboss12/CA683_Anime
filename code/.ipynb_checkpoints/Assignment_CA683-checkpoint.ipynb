- We need to remove records with a price of 0, as this is likely a data entry error.
- The Quantity and Price variables are highly skewed to the right, as indicated by the summary statistics.
- To deal with this skewness, we will follow Chen's (2012) recommendation in "Data Mining for the Online Retail Industry" and remove the top 1% of the most extreme values in each variable.

```python
# Remove records where price = 0
data = data[data["UnitPrice"] != 0]

# Remove top 1% of rare quantity and price values
quantity_threshold = data["Quantity"].quantile(0.99)
price_threshold = data["UnitPrice"].quantile(0.99)

data = data[data["Quantity"] < quantity_threshold]
data = data[data["UnitPrice"] < price_threshold]

# Print the new shape of the dataframe
print(f"New dataframe shape: {data.shape}")
```

## Generating advanced RFMT figures

This code generates advanced RFMT figures for customer segmentation by retrieving Recency, Frequency, Monetary, and Tenure data from the original dataset. It defines the max date in the data as the reference point for calculating recency and tenure, and creates new columns for purchase interval time and total spent. It then groups the data by customer ID and applies lambda functions to calculate recency, frequency, monetary, and tenure. The code also renames the columns for clarity and calculates the average monetary value per transaction. Finally, it removes any customers with a monetary value of 0.

```python
# Retrieve Recency, Frequency, Monetary, and Tenure
# Define the max date in the data as the reference point for calculating recency and tenure
df_date = data["InvoiceDate"].max()

# Create a new column for purchase interval time, which is the time difference between the purchase date and the reference point
data["PurchaseIntervalTime"] = data["InvoiceDate"]

# Create a new column for total spent, which is the product of the unit price and the quantity purchased
data["TotalSpent"]= data["UnitPrice"] * data["Quantity"]

# Define a dictionary of lambda functions to apply to each column in the groupby aggregation
new_advanced_field = {
    "InvoiceDate":lambda date:(df_date - date.max()).days,
    "InvoiceNo":lambda num: num.nunique(),
    "TotalSpent": lambda price:price.sum(),
    "PurchaseIntervalTime":lambda date: (df_date - date.min()).days
}
```

```python
data.columns
```

```python
# Group the data by customer ID and apply the lambda functions to calculate recency, frequency, monetary, and tenure
rfmt_df = data.groupby("CustomerID").agg(new_advanced_field)

# Rename the columns for clarity
rfmt_df.columns = ["Recency", "Frequency", "Monetary", "Tenure"]

# Calculate the average monetary value per transaction (i.e. Monetary/Frequency)
rfmt_df["Monetary"] = rfmt_df["Monetary"] / rfmt_df["Frequency"]

# Remove any customers with a monetary value of 0
rfmt_df = rfmt_df[rfmt_df["Monetary"] != 0]
```

## Apply the K-means++ clustering algorithm to create customer segments based on the PCA components

This code performs the elbow method to determine the optimal number of clusters (k) for KMeans clustering algorithm.

The algorithm is executed for k values ranging from 1 to 10, and the sum of squared distances (SSE) of samples to their closest cluster center is calculated for each k.

Then, a line plot is created with k on the x-axis and SSE on the y-axis to visualize the relationship between k and SSE. The point where the decrease in SSE slows down is often referred to as the "elbow", and it indicates the optimal number of clusters. The plot can be used to identify the number of clusters that strikes a balance between minimizing the SSE and avoiding overfitting.

```python
# Fit KMeans and calculate SSE for each *k*
sse = {}
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(data_norm)
    sse[k] = kmeans.inertia_
    
# Plot SSE for each *k*
plt.title('The Elbow Method')
plt.xlabel('k'); plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show();
```

This code performs KMeans clustering on the PCA-transformed data for different numbers of clusters and plots the silhouette analysis to determine the optimal number of clusters.

The code loops through a list of values for the number of clusters and for each value, it creates a subplot with two columns - the first column displays the silhouette plot and the second column displays a scatter plot of the PCA data with the clusters visualized.

For each value of the number of clusters, the KMeans algorithm is fit to the PCA-transformed data, and the cluster labels are computed. The silhouette score and sample silhouette values are then calculated using the silhouette_score and silhouette_samples functions from scikit-learn. These values are used to plot the silhouette analysis for each cluster, which shows the silhouette coefficient values for each sample in each cluster, as well as the average silhouette score for all clusters.

The scatter plot in the second column shows the PCA data points colored by their cluster label, with the cluster centers also displayed. The optimal number of clusters can be determined by finding the value of k that maximizes the average silhouette score.

```python
range_n_clusters = [2, 3, 4]

X = PCA_components

for n_clusters in range_n_clusters:
    
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    ax1.set_xlim([-0.1, 1])
    # The (n_clusters+1)*10 is for inserting blank space between silhouette
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value and a random generator
    # and a random seed for reproducibility.
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(X)

    # The silhouette_score gives the average value for all the samples.
    silhouette_avg = metrics.silhouette_score(X, cluster_labels)
    print(
        "For n_clusters =",
        n_clusters,
        "The average silhouette_score is :",
        silhouette_avg,
    )

    # Compute the silhouette scores for each sample
    sample_silhouette_values = metrics.silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to
        # each range of cluster 'i'
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(
            np.arange(y_lower, y_upper),
            0,
            ith_cluster_silhouette_values,
            facecolor=color,
            edgecolor=color,
            alpha=0.7,
        )

        # Label the silhouette plots with their cluster numbers
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the yaxis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(
        X.iloc[:, 0], X.iloc[:, 1], marker=".", s=30, lw=0, alpha=0.7, c=colors, edgecolor="k"
    )

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    # Draw white circles at cluster centers
    ax2.scatter(
        centers[:, 0],
        centers[:, 1],
        marker="o",
        c="white",
        alpha=1,
        s=200,
        edgecolor="k",
    )

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker="$%d$" % i, alpha=1, s=50, edgecolor="k")

    ax2.set_title("The visualization of the PCA data.")
    ax2.set_xlabel("Feature space for the PCA_1")
    ax2.set_ylabel("Feature space for the PCA_2")

    plt.suptitle(
        "Silhouette analysis for KMeans clustering on sample data with n_clusters = %d"
        % n_clusters,
        fontsize=14,
        fontweight="bold",
    )

plt.show()
```

```python
# Choose k=3 and fit data set to k-means model
X = data_norm
kmeans = KMeans(n_clusters=3, random_state=1)
kmeans.fit(X)

# Assign k-means labels to cluster labels
cluster_labels = kmeans.labels_

# Assign cluster labels to original pre-transformed data set
data_k3 = rfmt_df.assign(Cluster = cluster_labels)

# Group data set by k-means cluster
cluster_data = data_k3.groupby(['Cluster']).agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': ['mean'],
    'Tenure': ['mean', 'count']
}).round(0)

cluster_data.columns = ['Recency', 'Frequency', 'Monetary', 'Tenure', 'Size']
cluster_data
```

This code is adding new columns to the "cluster_data" dataframe based on certain conditions.

The first line sets a threshold for long-term users based on the number of months they have been using the service.

The second line sets a threshold for lost users based on the number of days since their last interaction with the service.

The third line sets a threshold for valuable users based on their frequency and monetary value, using the quantile function to select the top 50% of users in terms of these two variables.

The fourth line calculates the percentage of users in each cluster based on the total size of the cluster.

Finally, the code creates a new dataframe with the columns "Longterm", "Lost", "Valuable", and "Percentage" and assigns it to the original "cluster_data" dataframe. The "Longterm" column is True for users who have been using the service for more than the threshold_T number of months, the "Lost" column is True for users who have not interacted with the service in more than threshold_R number of days, and the "Valuable" column is True for users whose frequency and monetary value are both above the threshold_F_M value. The "Percentage" column contains the percentage of users in each cluster based on the total size of the cluster.

```python
threshold_T = 30*6 
threshold_R = 30
threshold_F_M = 0.5 

# long-term user, valuable user, lost user 
cluster_data['Longterm'] = cluster_data['Tenure'] > threshold_T
cluster_data['Lost'] = cluster_data['Recency'] >= threshold_R
cluster_data['Valuable'] = (cluster_data['Frequency'] >= cluster_data['Frequency'].quantile(threshold_F_M)) & (cluster_data['Monetary'] >= cluster_data['Monetary'].quantile(threshold_F_M))
cluster_data['Percentage'] = cluster_data['Size'] / cluster_data['Size'].sum() * 100
cluster_data[['Longterm', 'Lost', 'Valuable', 'Percentage']]
```

### Cluster explaination 

This output shows a table with four columns: Longterm, Lost, Valuable, and Percentage. The rows represent different clusters. The values in the columns indicate whether a customer belongs to a specific category or not based on certain thresholds.

- **Longterm**: Indicates whether a customer is a long-term user or not. The threshold for being considered a long-term user is threshold_T = 30x6 which is equivalent to 180 days (6 months) in this case. A customer is considered a long-term user if their Tenure (the time since they first became a customer) is greater than 180 days.
- **Lost**: Indicates whether a customer is a lost user or not. The threshold for being considered a lost user is threshold_R = 30, which means that a customer is considered lost if their Recency (the time since their last purchase) is greater than or equal to 30 days.
- **Valuable**: Indicates whether a customer is a valuable user or not. A customer is considered valuable if their Frequency (the number of purchases they have made) is greater than or equal to the 50th percentile of all customers' frequencies (threshold_F_M = 0.5), and their Monetary value (the total amount they have spent) is greater than or equal to the 50th percentile of all customers' monetary values.
- **Percentage**: Indicates the percentage of customers in each cluster relative to the total number of customers.

- **Cluster 1** contains customers who are valuable and lost, but not necessarily long-term. These customers represent 33.4% of the total customer base.

- **Cluster 2** contains customers who are long-term and lost, but not necessarily valuable. These customers represent 37.7% of the total customer base.
